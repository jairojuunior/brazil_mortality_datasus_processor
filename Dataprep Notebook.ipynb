## Dataprep

Please refer to https://icd.codes/icd10cm for information on ICD10 chapter codes.

```python
import numpy as np 
import pandas as pd
import io
import os
```

```python
for file in os.listdir("raw_files"):
    #File year identification
    year=""
    fp = io.open("raw_files/"+file, mode="r", encoding="cp1258")
    for i, line in enumerate(fp):
        if i==2:
            year = line[8:-1]
            break
    fp.close()
    print("Year: "+year)

    #Import
    df = pd.read_csv("raw_files/"+file, 
                        sep=";", decimal=".", skiprows=3, 
                        quotechar='"', encoding="cp1258")

    #Separate the city IBGE code
    df['cod_municipio'] = df['Município'].apply(lambda x: x[0:6])

    #Convert to numeric 
    for col in df.columns:
        if col not in ["Município", "cod_municipio"]:
            if df[col].dtype == "object":
                try:
                    df[col] = pd.to_numeric(df[col])
                except:
                    df[col] = pd.to_numeric(df[col].apply(lambda x: str(x).replace("-","0")), errors="coerce")
    
    #Normalize columns names according to ICD10
    
    for col_index in list(range(len(df.columns))):
        col = df.columns[col_index]
        if col[:3].lower() == "cap":
            df.rename(columns={col: "ICD10 Chapter "+col[4:]   }, inplace=True)
    
    #Operation extract
    print(df.info())
                
    #Output result
    df.to_csv("output_mortality/brazil_city_mortality_"+year+".csv", 
              index=False, sep=";", encoding="utf-8")
```
